```python
import pandas as pd
```

```python
orders = pd.read_csv('Orders.csv',header=0 ,index_col=0)
orders
```

```python
customers = pd.read_csv('Customers.csv',header=0,index_col=0)
customers
```

```python
cust_order = pd.merge(customers,orders,on='Customer_ID')
cust_order
```

```python
cust_order_opertion = cust_order[cust_order['Amount']>500].groupby('Country').agg(
    {
        'Amount':['sum'],
        'Customer_Name':['count']
    }
)
cust_order_opertion
```

```python
cust_order_sort = cust_order.groupby('Customer_Name').agg({'Amount':'sum'}).sort_values(by='Amount',ascending=False)
cust_order_sort

```
